Question 1:

What is Simple Linear Regression?

ANSWER

Simple Linear Regression is a statistical and machine learning technique used to find the relationship between one independent variable (X) and one dependent variable (Y). It predicts the value of the dependent variable by fitting a straight line to the data.

The equation of Simple Linear Regression is:

  Y=mX+c

  Where:

* Y = Dependent variable (output)

* X = Independent variable (input)

* m = Slope of the line

* c = Intercept (value of Y when X = 0)

The regression line is chosen so that it best represents the relationship between X and Y and helps predict future values.

Question 2: What are the key assumptions of Simple Linear Regression?


ANSWER:

Simple Linear Regression is based on the following key assumptions:

1. Linear Relationship: There should be a straight-line relationship between the independent variable (X) and the dependent variable (Y).

2. Independence of Observations: Each data point should be independent of the others.

3. Constant Variance (Homoscedasticity): The errors (residuals) should have approximately the same variance for all values of X.

4. Normal Distribution of Errors: The residuals should be normally distributed.

5. No Significant Outliers: The dataset should not contain extreme outliers that strongly affect the regression line.

Q3. Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

ANSWER:

* Heteroscedasticity is a condition in which the variance of the errors (residuals) is not constant across different values of the independent variable. In other words, the spread of the residuals changes as the value of X changes.

* It is important to address heteroscedasticity because regression assumes that the error variance remains constant. If this assumption is violated, the model's results and statistical inferences may become unreliable.

Question 4: What is Multiple Linear Regression?

ANSWER:

Multiple Linear Regression (MLR) is a statistical and machine learning technique used to study the relationship between one dependent variable (Y) and two or more independent variables (X₁, X₂, X₃, ...). It predicts the value of the dependent variable by fitting a linear equation using multiple input variables.

The equation of Multiple Linear Regression are:

Y=β0​+β1​X1​+β2​X2​+⋯+βn​Xn​

Where:

* Y = Dependent variable (output)

* X₁, X₂, ..., Xₙ = Independent variables (inputs)

* β₀ = Intercept

* β₁, β₂, ..., βₙ = Regression coefficients

Question 5: What is polynomial regression, and how does it differ from linear
regression?

ANSWER:

Polynomial Regression is a regression technique used when the relationship between the independent variable (X) and the dependent variable (Y) is non-linear. It fits a curved (polynomial) equation to the data instead of a straight line.

A polynomial regression equation is:

Y=β0​+β1​X+β2​X2+⋯+βn​Xn

| **Linear Regression**                              | **Polynomial Regression**                                                    |
| -------------------------------------------------- | ---------------------------------------------------------------------------- |
| Fits a **straight line** to the data.              | Fits a **curved line** to the data.                                          |
| Assumes a **linear relationship** between X and Y. | Models a **non-linear relationship** between X and Y using polynomial terms. |


```python
#Q6. Question 6: Implement a Python program to fit a Simple Linear Regression model to
# the following sample data:
# X = [1, 2, 3, 4, 5]
# Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points.
# (Include your Python code and output in the code box below.)?

#ANSWER

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print model parameters
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

# Plot data points
plt.scatter(X, Y, color="blue", label="Data Points")

# Plot regression line
plt.plot(X, Y_pred, color="red", label="Regression Line")

# Labels and title
plt.title("Simple Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Show plot
plt.show()
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
# Area = [1200, 1500, 1800, 2000]
# Rooms = [2, 3, 3, 4]
# Price = [250000, 300000, 320000, 370000]
# Check for multicollinearity using VIF and report the results.
# (Include your Python code and output in the code box below.)?

#ANSWER

# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create the dataset
data = pd.DataFrame({
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
})

# Independent and dependent variables
X = data[["Area", "Rooms"]]
Y = data["Price"]

# Train Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print model parameters
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

```python
#uestion 8: Implement polynomial regression on the following data:
# X = [1, 2, 3, 4, 5]
# 3
# Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.
# (Include your Python code and output in the code box below.)?

#ANSWER

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (Degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Print model parameters
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Plot original data
plt.scatter(X, Y, color="blue", label="Data Points")

# Plot polynomial regression curve
plt.plot(X, Y_pred, color="red", linewidth=2, label="Polynomial Curve")

# Labels and title
plt.title("Polynomial Regression (Degree = 2)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Display graph
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
# X = [10, 20, 30, 40, 50]
# Y = [15, 35, 40, 50, 65]
# Assess heteroscedasticity by examining the spread of residuals.
# (Include your Python code and output in the code box below.)?

#ANSWER

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print residual values
print("Residuals:")
print(residuals)

# Create residual plot
plt.scatter(Y_pred, residuals, color="blue")
plt.axhline(y=0, color="red", linestyle="--")

# Labels and title
plt.title("Residuals Plot")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")

# Show plot
plt.show()
```

Question 10: Imagine you are a data scientist working for a real estate company.
You need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model?

ANSWER

To build a reliable house price prediction model, the following steps should be taken:

1. Detect heteroscedasticity by examining the residual plot. If the spread of residuals is not constant, it indicates heteroscedasticity.

2. Reduce heteroscedasticity by applying a suitable transformation (such as the logarithm of the target variable) or using appropriate regression techniques so that the error variance becomes more stable.

3. Detect multicollinearity by calculating the Variance Inflation Factor (VIF) for the independent variables.

4. Reduce multicollinearity by removing or combining highly correlated features, or selecting only the most relevant variables.

5. Retrain and evaluate the regression model after addressing both issues to ensure the model produces reliable predictions.
